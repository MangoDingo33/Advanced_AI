```python
import pandas as pd
```

```python
data = pd.read_csv("dataset/car.data", header=None)
df = pd.DataFrame(data)
```

```python
df
```

```python
df.columns =['buying', 
            'maint',
            'doors',
            'persons',
            'lug_boot',
            'safety',
            'Class']
```

```python
df
```

```python
#인코딩
df = pd.get_dummies(df, columns =['buying', 'maint', 'doors', 'persons','lug_boot', 'safety'])
```

```python
df
```

```python
#라벨링

from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()

df['buying_high']=label.fit_transform(df['buying_high'])
df['buying_low']=label.fit_transform(df['buying_low'])
df['buying_med']=label.fit_transform(df['buying_med'])
df['buying_vhigh']=label.fit_transform(df['buying_vhigh'])
df['maint_high']=label.fit_transform(df['maint_high'])
df['maint_low']=label.fit_transform(df['maint_low'])
df['maint_med']=label.fit_transform(df['maint_med'])
df['maint_vhigh']=label.fit_transform(df['maint_vhigh'])
df['doors_2']=label.fit_transform(df['doors_2'])
df['doors_3']=label.fit_transform(df['doors_3'])
df['doors_4']=label.fit_transform(df['doors_4'])
df['doors_5more']=label.fit_transform(df['doors_5more'])
df['persons_2']=label.fit_transform(df['persons_2'])
df['persons_4']=label.fit_transform(df['persons_4'])
df['persons_more']=label.fit_transform(df['persons_more'])
df['lug_boot_big']=label.fit_transform(df['lug_boot_big'])
df['lug_boot_med']=label.fit_transform(df['lug_boot_med'])
df['lug_boot_small']=label.fit_transform(df['lug_boot_small'])
df['safety_high']=label.fit_transform(df['safety_high'])
df['safety_low']=label.fit_transform(df['safety_low'])
df['safety_med']=label.fit_transform(df['safety_med'])

df
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['Class'], axis=1)
y = df['Class']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# DT

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
# SVM

from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

pred_svm = clf_svm.predict(x_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
# LR

from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
# RF

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)
pred = rf_clf.predict(x_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
